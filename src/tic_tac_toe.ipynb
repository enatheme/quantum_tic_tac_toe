```python
from qiskit.visualization import plot_histogram
%matplotlib inline
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (10,10)
```

```python
import tic_tac_toe as ttt
```

```python
game = ttt.q_ttt(['1', '0', '0', '1', ' ', ' ', '0', ' ', ' '], to_gate = True)
```

```python
ttt.display(game)
```

```python
job = ttt.run_it(game, local_simulator = False)
```

```python
# job id 622dbb38817d61cd61a88a84
# expected |0011> |1001>
plot_histogram(job.result().get_counts(), title="first run")
```

```python
game2 = ttt.q_ttt(['1', '0', '0', '1', ' ', ' ', ' ', ' ', ' '], to_gate = True)
job2 = ttt.run_it(game2, local_simulator = False)
```

```python
# job id 622dbbe26858d0d41ec13857
# expected: |???11> |xx1xx> |1xxx1>
# reduce the number of items
r = job2.result().get_counts()
reduced = {}
for i in r:
    if r[i] > 20:
        reduced[i] = r[i]
reduced
plot_histogram(reduced, title="second run")
```

```python
game3 = ttt.q_ttt(['1', '0', '0', '1', ' ', ' ', '0', '0', '1'], to_gate = True)
job3 = ttt.run_it(game3, local_simulator = False)
```

```python
# job id 622dbc5b317810f5c92047d2
# expected: |01>
plot_histogram(job3.result().get_counts(), title="third run")
```
